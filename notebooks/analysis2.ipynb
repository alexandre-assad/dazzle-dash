```python
from os import path
import pandas as pd 
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

from importnb import Notebook
import plotly.io as pio
```

```python
df = Notebook.load_file(path.join("exploratory_analysis.ipynb")).df
pio.renderers.default = 'notebook_connected'
```

```python
def get_year_production(liste: list, index: int) -> int:
    return liste[index]

def sum_production(liste: list) -> int:
    return sum(liste)
```

```python
df['Total'] = df["years_evolution"].apply(sum_production)
item = "Sugar cane"

plt.figure(figsize=(20,10))  # Augmenter la taille de la figure

# Filtrer les donnÃ©es pour l'aliment actuel
df_item = df[df['Item'] == item]
three_best_producers = df_item.sort_values(["Total"], ascending=False).iloc[:3]
years = [i for i in range(1961, 2014)]
# # Parcourir chaque pays
for _, country in three_best_producers.iterrows():
    # Tracer la courbe pour le pays actuel
    plt.plot(years, country["years_evolution"])

plt.title(f'Production de {item} dans tous les pays de 1961 Ã  2013', fontsize=14)
plt.legend(three_best_producers["Area"])
plt.xlabel('AnnÃ©e', fontsize=12)
plt.ylabel('Production totale', fontsize=12)
plt.xticks(rotation=90, fontsize=8)
plt.grid(axis='y')
plt.tight_layout()  # Ajuster l'espacement

# Afficher le graphique
plt.show()


# Liste des 
```
